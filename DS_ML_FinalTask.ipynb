```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('student-math.csv', sep = ';')
data
```

```python
data['Final_grade'] = data.G1 + data.G2 + data.G3
```

```python
data
```

```python
y = np.array(data['Final_grade'])
y
```

```python
X = data.drop(['G3', 'Final_grade'], axis = 'columns')
```

```python
# New data sheet which does not contain column G3
X
```

##### By Using Label Encoder

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_school = LabelEncoder()
le_sex = LabelEncoder()
le_age = LabelEncoder()
le_address = LabelEncoder()
le_famsize = LabelEncoder()
le_Pstatus = LabelEncoder()
le_Mjob = LabelEncoder()
le_Fjob = LabelEncoder()
le_reason = LabelEncoder()
le_guardian = LabelEncoder()
le_schoolsup = LabelEncoder()
le_famsup = LabelEncoder()
le_paid = LabelEncoder()
le_activities = LabelEncoder()
le_nursery = LabelEncoder()
le_higher = LabelEncoder()
le_internet = LabelEncoder()
le_romantic = LabelEncoder()
```

```python
X['school_n'] = le_school.fit_transform(X['school'])
X['sex_n'] = le_sex.fit_transform(X['sex'])
X['age_n'] = le_age.fit_transform(X['age'])
X['address_n'] = le_address.fit_transform(X['address'])
X['famsize_n'] = le_famsize.fit_transform(X['famsize'])
X['Pstatus_n'] = le_Pstatus.fit_transform(X['Pstatus'])
X['Mjob_n'] = le_Mjob.fit_transform(X['Mjob'])
X['Fjob_n'] = le_Fjob.fit_transform(X['Fjob'])
X['reason_n'] = le_reason.fit_transform(X['reason'])
X['guardian_n'] = le_guardian.fit_transform(X['guardian'])
X['schoolsup_n'] = le_schoolsup.fit_transform(X['schoolsup'])
X['famsup_n'] = le_famsup.fit_transform(X['famsup'])
X['paid_n'] = le_paid.fit_transform(X['paid'])
X['activities_n'] = le_activities.fit_transform(X['activities'])
X['nursery_n'] = le_nursery.fit_transform(X['nursery'])
X['higher_n'] = le_higher.fit_transform(X['higher'])
X['internet_n'] = le_internet.fit_transform(X['internet'])
X['romantic_n'] = le_romantic.fit_transform(X['romantic'])
```

```python
X
```

```python
# Drop the unrequired columns
X_New = X.drop(['school','sex','age','address','famsize','Pstatus','Mjob','Fjob','reason','guardian','schoolsup','famsup','paid','activities','nursery','higher','internet','romantic'], axis = 'columns')
```

##### Final dataset

```python
X_New
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_New, y, test_size = 0.2)
```

```python
len(X_train)
```

```python
len(X_test)
```

##### By Using Linear Regression

```python
from sklearn import linear_model
from sklearn.linear_model import LinearRegression
```

```python
reg = linear_model.LinearRegression()
```

```python
reg.fit(X_New, y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
y_predicted = reg.predict(X_test)
y_predicted
```
