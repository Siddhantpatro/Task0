```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("student-math.csv", sep = ";")
df
```

```python
# sum of the columns G1, G2, G3 and is stored in Final_grade

df['Final_grade'] = df.G1.values + df.G2.values + df.G3.values  
df
```

```python
# G1,G2,G3 is dropped

df_New = df.drop(['G1', 'G2', 'G3'], axis = 1)
df_New
```

##### Binary values in the table is converted to 1 & 0

```python
dfNew1 = df[['nursery', 'higher', 'romantic', 'internet', 'activities', 'paid', 'famsup', 'schoolsup' ]].replace({'yes': 1 , 'no': 0})
dfNew1
```

```python
df_New[['schoolsup', 'famsup', 'paid', 'activities', 'nursery', 'higher', 'internet', 'romantic' ]] = dfNew1
df_New
```

##### For Address

```python
df_Newaddress = df[['address']].replace({'U': 1 , 'R': 0})
df_Newaddress
```

```python
df_New[['address']] = df_Newaddress
df_New[['address']]
```

##### School Table

```python
df_Newschool = df[['school']].replace({'GP': 1 , 'MS': 0})
df_Newschool
```

```python
df_New[['school']] = df_Newschool
df_New[['school']]
```

##### Famsize Table

```python
df_Newfamsize = df[['famsize']].replace({'GT3': 1 , 'LE3': 0})
df_Newfamsize
```

```python
df[['famsize']] = df_Newfamsize
df[['famsize']]
```

##### Pstatus Table

```python
df_New.Pstatus = df[['Pstatus']].replace({'T': 1 , 'A': 0})
df_New.Pstatus
```

```python
df_New[['Pstatus']].replace(df_New.Pstatus)
```

##### Sex Table

```python
df_Newsex = df[['sex']].replace({'M': 1 , 'F': 0})
df_Newsex
```

```python
df_New[['sex']] = df_Newsex
df_New[['sex']]
```

##### Graph between studytime and final_grade

```python
df_New[['studytime', 'Final_grade']]
```

```python
Study_time = df_New['studytime']
Final_grade = df_New['Final_grade']
```

```python
plt.scatter(Study_time, Final_grade, marker = '.', color = 'red')

plt.xlabel('Study_time')
plt.ylabel('Final_grade')

plt.title('Study time vs Final grade')

plt.grid(True)


plt.style.use('fivethirtyeight')
```

```python
box_plot_data=[Final_grade, Study_time]

plt.boxplot(box_plot_data,patch_artist=True,labels=['Final_grade','Study_time'])

plt.grid(True)

plt.show()

plt.style.use('fivethirtyeight')
```
